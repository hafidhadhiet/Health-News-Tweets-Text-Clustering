<p>Selanjutnya kami menyeleksi kata-kata yang tergolong stopwords yang terdapat dalam wordcloud yang sudah ditampilkan. Stopwords adalah kata-kata yang sekiranya tidak memiliki relevansi terhadap topik yang sedang dikerjakan. Banyak sekali kata-kata random yang tidak memiliki arti yang akan sangat memengaruhi proses clustering. Selain itu karena tweet dalam dataset ini ditulis dalam Bahasa Inggris, kami juga menggunakan package ENGLISH_STOP_WORDS dari library sklearn.feature_extraction. Package ini berisi stopwords yang umum ada dalam bahasa inggris seperti "the", "and", "or", dll yang tidak ada relevansinya terhadap suatu topik khusus. Selanjutnya kami akan menghapus kata-kata yang tergolong stopwords tersebut dari dataset kami.</p>

```python
stopwords = set(STOPWORDS)
stopwords.update(['bit','ly','n','pr','reut','rs','cnn','lat','ms',
                 'nbcnews','cmp','rss','p2yZYB','msn','wp','jpg','VIDEO','wsj',
                 't','co','try','pbs','twimg','gu','RT','stefaniei','ws','OW','cbc',
                 'cbc','ca','bbc','U','S','ca','khne','WS','NH','amp','lati','nhs',
                 'stw','theguardian','https','tw','ow','gt','p2yzyb','rt','html',
                  'nhs','trib','1gny60o','gdnhealthcare','q1','q2','q3','q4','q5','q6',
                 '1ajrqyn', 'hkmaub','npf','1ajrirw','q7','1akyupf','1gnn3nv'])

my_stopwords = ENGLISH_STOP_WORDS.union(stopwords)

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

<p>Selanjutnya kami lakukan encoding dataset tersebut menjadi ke dalam bentuk unicode. Unicode merupakan bentuk standar encoding untuk merepresentasikan suatu text secara digital. Dilakukan encoding ke dalam unicode agar dapat dilakukan clustering.</p>

```python
dataset = df['Tweet'].values.astype('U')
```

<p>Selanjutnya kami menggunakan package TfidfVectorizer dari library sklearn.feature_extraction. Package ini melakukan penghapusan kata-kata yang tergolong stopwords yang tadi sudah kami definisikan</p>

```python
vectorizer = TfidfVectorizer(stop_words=my_stopwords)
features = vectorizer.fit_transform(dataset)
```

<h3>Proses Clustering</h3>
<p>Selanjutnya kami akan melakukan proses clustering menggunakan KMeans. Kami menggunakan KMeans karena dataset yang kami miliki sangat kompleks yang berisi hampir 6 juta kata yang unique/berbeda. Sebelum melakukan proses clustering, kami mencari nilai k optimal terlebih dahulu menggunakan metode elbow.</p>

```python
distortions = []
K = range(1, 21)
for k in K:
    kmeanmodel = KMeans(n_clusters=k, max_iter=100)
    kmeanmodel.fit(features)
    distortions.append(kmeanmodel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Elbow Method to show optimal k')
plt.show()
```

<p>Grafik tidak menunjukkan pola elbow bahkan sampai nilai k=20. Karena keterbatasan hardware yang kami miliki yang tidak memungkinkan kami untuk melakukan percobaan Elbow Method dengan nilai k>20 karena prosesnya akan sangat lama, maka pada tugas kali ini, kami akan menetapkan nilai k=20.</p>

```python
k = 20
model = KMeans(n_clusters=k)
model.fit(features)
```

```python
df['Cluster'] = model.labels_
df.head()
```

```python
clusters = df.groupby('Cluster')

for cluster in clusters.groups:
#     f = open('cluster'+str(cluster)+ '.csv', 'w')
    data = clusters.get_group(cluster)[['Tweet']]
#     f.write(data.to_csv(index_label='id'))
#     f.close()
data
```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]:
        print(' %s' % terms[j])
    print('-------------')
```
